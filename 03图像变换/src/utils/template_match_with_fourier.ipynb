```python
import numpy as np
from PIL import Image
import cv2
from matplotlib import pyplot as plt
```

```python
bw = Image.open(r"H:\编程工程\数字图像处理\实验四\数字图像处理实验4\imgs\Lena.jpg").convert("L")
bw = np.array(bw)
plt.imshow(bw,cmap="gray")
plt.show()
```

```python
eye = Image.open(r"H:\编程工程\数字图像处理\实验四\数字图像处理实验4\imgs\Lena_eye.bmp").convert("L")
eye = np.array(eye)
plt.imshow(eye,cmap="gray")
plt.show()
```

```python
eye_ = np.zeros([300,300])
eye_[:eye.shape[0],:eye.shape[1]] = eye
```

```python
C = np.real(np.fft.ifft2(np.fft.fft2(bw) *
    np.fft.fft2(np.rot90(eye_,2),(300,300))))
C = map(C,255,0)
plt.imshow(C,cmap="gray")
plt.show()
```

```python
def map(arr,max,min):
    x_min = np.min(arr)
    x_max = np.max(arr)
    return min + (max - min)/(x_max - x_min) * (arr - x_min)
```

```python
thresh = 250
plt.imshow(C>thresh,cmap="gray")
plt.show()
```
