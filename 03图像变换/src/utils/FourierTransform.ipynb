```python
import numpy as np
from PIL import Image
import cv2
from matplotlib import pyplot as plt
```

```python
img = Image.open(r"F:\编程工程\数字图像处理\实验四\数字图像处理实验4\imgs\test.jpg").convert("L")
img = img.rotate(0)
plt.imshow(img)
```

```python
img = np.array(img)
f = np.fft.fft2(img)
```

```python
fshift = np.fft.fftshift(f) 

#fft结果是复数, 其绝对值结果是振幅
fimg = np.log(np.abs(fshift))
```

```python
plt.subplot(121)
plt.imshow(img,"gray")
plt.subplot(122)
plt.imshow(fimg,"gray")
```

```python
plt.imsave("fimg.jpg",fimg,cmap="gray")
```

```python
def test(img,angle):
    img = img.rotate(angle)
    img = np.array(img)
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f) 

    #fft结果是复数, 其绝对值结果是振幅
    fimg = np.log(np.abs(fshift))
    
    plt.subplot(121)
    plt.imshow(img,"gray")
    plt.subplot(122)
    plt.imshow(fimg,"gray")
```

```python
img = Image.open(r"F:\编程工程\数字图像处理\实验四\数字图像处理实验4\imgs\test.jpg").convert("L")
test(img,0)
```
