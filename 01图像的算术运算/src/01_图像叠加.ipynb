```python
%matplotlib inline
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
```

```python
im_1 = np.array(Image.open(r'数字图像处理\cyber.jpg'))
im_2 = np.array(Image.open(r'数字图像处理\Lena.jpg'))
```

```python
np.max(im_1)
np.min(im_1)
np.max(im_2)
np.min(im_2)
```

```python
im_1.shape
im_2.shape
```

```python
im_2_filled_with_0 = np.zeros(im_1.shape)
```

```python
im_2_filled_with_0[0:im_2.shape[0],0:im_2.shape[1],:] = im_2
```

```python
np.max(im_2_filled_with_0)
np.min(im_2_filled_with_0)
```

```python
mix_im = 0.5 * im_1 + 0.5 * im_2_filled_with_0
mix_im = mix_im.astype(np.uint8)
mix_im.dtype
```

```python
im = Image.fromarray(mix_im)
```

```python
plt.imshow(im)
```
